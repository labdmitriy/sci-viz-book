### Exercise 1

```python
point_size = 1 / 72
markers_count = 8
marker_size_data = 1

fig, ax = plt.subplots(figsize=(10, 3))
ax.set_xlim(0, marker_size_data * markers_count)
ax.set_ylim(0, marker_size_data + 1)
ax.set_aspect(1)

bbox = ax.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
width, height = bbox.width, bbox.height

data_unit_points = (width / markers_count) / point_size
print(data_unit_points.round(2))

for i in range(markers_count):
    ax.scatter(marker_size_data * i + marker_size_data / 2, 
               (marker_size_data + 1) / 2,
               s=data_unit_points**2, fc='none', ec='k')
plt.show()
```
